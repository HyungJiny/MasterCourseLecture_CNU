```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sec = []
reno = []
with open('./reno_output.csv', 'r') as reno_output:
    for line in reno_output.readlines():
        sec.append(float(line.strip().split(',')[0]))
        reno.append(int(line.strip().split(',')[-1]))

my_dpi = 94
plt.rcParams.update({'font.size': 13})
plt.figure(figsize=(800/my_dpi, 600/my_dpi), dpi=my_dpi)

plt.plot(reno)
plt.xlim(0, 200)

plt.savefig('./reno.png', bbox_inches='tight')
```
