```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sec = []
reno = []
with open('./reno_output.csv', 'r') as reno_output:
    for line in reno_output.readlines():
        sec.append(float(line.strip().split(',')[0]))
        reno.append(int(line.strip().split(',')[-1]))
plt.plot(reno)
plt.xlim(0, 1000)
```
